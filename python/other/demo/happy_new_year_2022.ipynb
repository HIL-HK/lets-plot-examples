```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def new_year_tree_layer():
    tree_data = {'y': [6] + 100 * ([5] + 2 * [4] + 3 * [3] + 4 * [2] + 5 * [1])}

    return geom_violin(aes(y='y'), data=tree_data, \
                       color='#00441b', fill='#41ab5d', bw=.3, \
                       tooltips='none')
```

```python
def star_layer():
    return geom_point(x=0, y=6, shape=8, size=15, color='#ef3b2c')
```

```python
def decorations_layer():
    import numpy as np

    np.random.seed(0)
    colors = ['#ef3b2c', '#54278f', '#9ecae1', '#fed976']
    shapes = [21, 23, 24]
    x = [-.09, -.26, .18, .33, -.03, .04, -.13, 0.11]
    y = [4.8, 2.8, 3.8, 1.8, 3.8, 2.7, 1.8, 1.4]
    decorations_data = {
        'x': x,
        'y': y,
        'shape': np.random.choice(shapes, size=len(x)),
        'color': np.random.choice(colors, size=len(x)),
    }

    return geom_point(aes('x', 'y', shape='shape', fill='color'), \
                      data=decorations_data, \
                      size=10, color='black') + \
               scale_fill_identity() + \
               scale_shape_identity()
```

```python
def greetings_layer():
    return geom_text(x=0, y=7, label='Happy 2022!', \
                     size=22, color='#4177a1', \
                     family='Times New Roman', fontface='bold')
```

```python
ggplot() + \
    new_year_tree_layer() + \
    star_layer() + \
    decorations_layer() + \
    greetings_layer() + \
    coord_fixed(ratio=.15) + \
    xlim(-.5, .5) + \
    theme_classic() + theme(axis='blank') + \
    ggsize(800, 600)
```
