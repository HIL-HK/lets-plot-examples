```python
from lets_plot.geo_data import *
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
country = 'Cuba'
bbox = geocode_countries(names=country)
gdf = geocode_cities().scope(bbox).get_centroids()

ggplot() + \
    geom_rect(data=bbox.get_boundaries(), fill='#95a5a6', size=0) + \
    geom_point(data=gdf, color='white', \
               tooltips=layer_tooltips().line('@{found name}')) + \
    ggsize(800, 600) + ggtitle(country) + \
    theme_void()
```
