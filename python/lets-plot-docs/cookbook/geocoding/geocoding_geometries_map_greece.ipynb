```python
country = 'Greece'
gdf = geocode_counties().scope(country).inc_res(4).get_boundaries()

ggplot() + \
    geom_map(data=gdf, color='white', fill='#1a242f', \
             tooltips=layer_tooltips().line('@{found name}')) + \
    ggsize(800, 600) + ggtitle(country) + \
    theme_void()
```
