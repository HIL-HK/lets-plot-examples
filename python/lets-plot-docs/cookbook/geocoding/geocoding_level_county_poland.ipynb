```python
from lets_plot.geo_data import *
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
country = 'Poland'
gdf = geocode_counties().scope(country).inc_res(4).get_boundaries()

ggplot() + \
    geom_map(data=gdf, color='white', fill='#18bc9c', \
             tooltips=layer_tooltips().line('@{found name}')) + \
    ggsize(800, 600) + ggtitle('{0} Counties'.format(country)) + \
    theme_void()
```
