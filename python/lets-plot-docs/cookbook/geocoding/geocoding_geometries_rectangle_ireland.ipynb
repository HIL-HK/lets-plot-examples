```python
from lets_plot.geo_data import *
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
country = 'Ireland'
country_gdf = geocode_countries(names=country).inc_res(4).get_boundaries()
gdf = geocode_counties().scope(country).get_boundaries()

ggplot() + \
    geom_map(data=country_gdf, fill='#95a5a6', size=0) + \
    geom_rect(data=gdf, color='#95a5a6', size=1, alpha=0) + \
    geom_rect(data=gdf, color='white', size=.5, alpha=0, \
              tooltips=layer_tooltips().line('@{found name}')) + \
    ggsize(800, 600) + ggtitle(country) + \
    theme_void()
```
