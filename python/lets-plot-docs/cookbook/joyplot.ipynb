# Joyplot

This example is taken from the ["Common Plots / Ridge, or ‘joy’, plots"](https://aeturrell.github.io/coding-for-economists/vis-common-plots.html#ridge-or-joy-plots) sub-section of the book [Coding for Economists](https://aeturrell.github.io/coding-for-economists/) (1st edition).

```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
df = pd.read_pickle("https://github.com/aeturrell/coding-for-economists/raw/main/data/berkeley_data.pkl")
df.head()
```

```python
final_year = df["Year"].max()
first_year = df["Year"].min()

breaks = [y for y in list(df.Year.unique()) if y % 10 == 0]
(
    ggplot(df, aes("Anomaly", "Year", fill="Year"))
    + geom_area_ridges(scale=20, alpha=1, size=0.2, trim=True, show_legend=False)
    + scale_y_continuous(breaks=breaks, trans="reverse")
    + scale_fill_viridis(option="inferno")
    + ggtitle(
        "Global daily temperature anomaly {0}-{1} \n(°C above 1951-80 average)".format(
            first_year, final_year
        )
    )
)
```
