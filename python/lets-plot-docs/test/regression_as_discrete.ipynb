```python
d = {
    'x': [0, 1, 2],
    'v': [3, 1, 2],
}

p = ggplot(d) + ggsize(300, 200)

p + geom_point(aes(x='x', color='v'))
```

```python
p + geom_point(aes(x='x', color=as_discrete('v')))
```

```python
p + geom_point(aes(x='x', color=as_discrete('v', levels=[1, 2, 3])))
```

```python
p + geom_point(aes(x='x', color=as_discrete('v', levels=[1, 2, 3], order=-1)))
```

```python
p + geom_point(aes(x='x', color=as_discrete('v', order=-1)))
```

```python
p + geom_point(aes(x='x', color=as_discrete('v', label='VVV')))
```

```python
p + geom_point(aes(x='x', color=as_discrete('v', levels=[1, 2, 3], order=-1, label='VVV')))
```

Grouping

```python
df = {
    'x': [0, 5, 10, 15],
    'y': [0, 5, 10, 15],
    'a': [0, 0, 1, 1],
    'c': ['a', 'a', 'b', 'b']
}

ggplot(df, aes(x='x', y='y')) + geom_line(aes(color=as_discrete('a')), size=3)
```

```python
df = {
    'displ': [11, 12, 13, 2, 3, 4, 5, 6, 7, 8, 9, 10],
    'hwy':   [27, 25, 23, 45, 43, 41, 39, 37, 35, 33, 31, 29],
    'cyl':   [8, 8, 8, 4, 4, 4, 5, 5, 5, 6, 6, 6]
}

ggplot(df, aes(x='displ', y='hwy')) \
    + geom_point(aes(color=as_discrete('cyl'))) \
    + geom_smooth(aes(color=as_discrete('cyl')), method='lm', size=1, se=False)
```
