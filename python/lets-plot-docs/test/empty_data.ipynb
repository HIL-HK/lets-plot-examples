```python
import numpy as np

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def plot_tests(layer_name, *, exceptions=[], layer=None):
    tests = {
        'control': ggplot({'x': [0, 1], 'y': [0, 1]}, aes('x', 'y')),
        'empty': ggplot({'x': [], 'y': []}, aes('x', 'y')),
        'all_nan': ggplot({'x': [np.nan, np.nan], 'y': [np.nan, np.nan]}, aes('x', 'y')),
        'group_nan': ggplot({'x': [0, 1, np.nan, np.nan], 'y': [0, 1, 0, 1], 'g': ["A", "A", "B", "B"]}, \
                            aes('x', 'y', color='g', fill='g')),
        'facet_nan': ggplot({'x': [0, 1, np.nan, np.nan], 'y': [0, 1, 0, 1], 'g': ["A", "A", "B", "B"]}, \
                            aes('x', 'y')) + \
            facet_grid(x='g'),
        'facet_cross': ggplot({'x': [0, 1, 0, 1], 'y': [0, 1, 0, 1], 'g1': ["A", "A", "B", "B"], 'g2': ["C", "C", "D", "D"]}, \
                              aes('x', 'y')) + \
            facet_grid(x='g1', y='g2'),
        'lim': ggplot({'x': [0, 1], 'y': [0, 1]}, aes('x', 'y')) + xlim(2, 3),
    }

    if layer is None:
        layer = globals()[layer_name]()

    if any(exceptions):
        print("EXCEPTED TESTS FOR {0}():\n  {1}\n".format(layer_name, '\n  '.join(exceptions)))

    return gggrid([
        test_spec + layer + ggtitle("{0}(): {1}".format(layer_name, test_name))
        for test_name, test_spec in tests.items()
        if test_name not in exceptions
    ], ncol=3)
```
