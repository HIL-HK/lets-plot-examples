```python
import pandas as pd
from lets_plot import *
from lets_plot.bistro.corr import *

LetsPlot.setup_html()
```

```python
mpg = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv').drop(columns=['Unnamed: 0'])

mpg.head()
```

```python
["displ-longer-label"]*5
```

```python
bounds=theme(
    plot_background=element_rect(color="red", size=2),
    panel_background=element_rect(color="blue", size=1), 
    legend_background=element_rect(color="blue", size=1))

p=corr_plot(mpg.corr(numeric_only=True)).tiles().build() + ggtitle("Tiles") + bounds
p + scale_y_discrete(labels=["displ-displ"]*5)
```

```python
p + ggsize(500, 500)
```

```python
p + theme(legend_position=[1, 0.5], legend_justification=[0, 0.5]) + ggsize(500, 250)
```

```python
housing_df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/Ames_house_prices_train.csv")
housing_df.shape
```

```python
# Tmp workaround
housing_corr = housing_df.corr(numeric_only=True)
```

```python
corr_plot(housing_corr).tiles(type='lower').palette_BrBG().build() + bounds
```

```python
p = (corr_plot(housing_corr, threshold=.5).tiles(diag=False).palette_BrBG().build() 
 # + ggtitle("Threshold: 0.5")
 # + ggsize(550, 400) 
 + bounds
)

gggrid([
    p,
    p + theme(axis_text_x=element_blank()),
    p + theme(axis_text_x=element_text(angle=0)),
    p + theme(axis_text_x=element_text(angle=20)),
    p + theme(axis_text_x=element_text(angle=45)),
    p + theme(axis_text_x=element_text(angle=75)),
    p + theme(axis_text_x=element_text(angle=90)),
], ncol=1) + ggsize(550, 410 * 6)
```
