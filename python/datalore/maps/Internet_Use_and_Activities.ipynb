# Internet Use and Activities in Europe

Frequency of internet access: daily

Unit of measure: Percentage of individuals

Year: 2016

```python
import pandas as pd

from lets_plot import *
from lets_plot.geo_data import *
LetsPlot.setup_html()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/internet_use_and_activities.csv")
df.head()
```

```python
boundaries_gpd = geocode_countries(df.country).ignore_not_found().get_boundaries('country')
```

```python
ggplot() + \
    geom_livemap(zoom=4, location=[10, 53]) + \
    geom_polygon(aes(fill='percent'), data=df, map=boundaries_gpd, map_join='country',
                 alpha=0.6, color='black', size=0.1,
                 tooltips=layer_tooltips().format('percent', '{}%')) + \
    scale_fill_gradient(high='magenta', format="{}%") + \
    ggsize(800, 600) + \
    ggtitle('Internet Use and Activities in Europe')
```
