```python
from lets_plot.geo_data import *
```

```python
# No level - autodetect, in case of ambiguity take state
geocode(names="New York").get_geocodes()
```

```python
# city
geocode(names='New York', level='city').get_geocodes()
```

```python
geocode(names='warwick', level='city').allow_ambiguous().get_geocodes()
```

```python
from shapely.geometry import box

geocode(names='warwick', level='city') \
    .where('warwick', scope=box(-71.488034, 41.685644, -71.367528, 41.744844)) \
    .get_geocodes()
```

```python
c = geocode(level="country")
c.get_geocodes()
```

```python
states = geocode(names='us-48')
states.get_geocodes()
```

```python
ggplot() + geom_point(map=states, color='black', size=10)
```

```python
ggplot() + geom_polygon(aes(fill='state'), data=states, map=states, map_join=('state'))
```
