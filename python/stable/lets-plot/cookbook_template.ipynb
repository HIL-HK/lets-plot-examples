```python
import numpy as np
import pandas as pd

from lets_plot import *
from lets_plot.geo_data import *
from lets_plot.bistro import *
```

```python
LetsPlot.setup_html()
```

> *For demo, if possible, use standard data from [lets-plot-docs](https://github.com/JetBrains/lets-plot-docs/tree/master/data) repository.*
>
> *For example:*

```python
df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv')
df.head(3)
```

> *The 1-st figure is usually built with a minimal set of required parameters.*
>
> *In many cases chapter header is not necessary.*

```python
ggplot(df) + geom_point(aes(x="displ", y="hwy"))
```

<a id="color"></a>
#### 1. `color` and `size`

```python
ggplot(df) + geom_point(aes(x="displ", y="hwy", color="drv", size="displ"))
```

<a id="shape"></a>
#### 2. `shape`

```python
ggplot(df) + geom_point(aes(x="displ", y="hwy", shape="drv"), size=8)
```

<a id="hollow-shape"></a>
##### 2.1 Switching to Hollow Shapes 

```python
ggplot(df) + geom_point(aes(x="displ", y="hwy", shape="drv"), size=8) + \
    scale_shape(solid=False)
```
