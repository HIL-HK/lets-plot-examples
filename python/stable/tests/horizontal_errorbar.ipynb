```python
LetsPlot.setup_html()
```

```python
data = {
    'x': [2000, 2020],
    'ymin': [1005, 1006],
    'ymax': [1001, 1002],
}
```

```python
ggplot(data) + geom_errorbar(aes(x='x', ymin='ymin', ymax='ymax'))
```

```python
ggplot(data) + geom_crossbar(aes(x='x', ymin='ymin', ymax='ymax'))
```

```python
ggplot({'x': [2000, 2000, 2000, 2000, 2020, 2020, 2020, 2020],
        'y': [1000, 1001, 1002, 1003, 1004, 1005, 1006, 1007]}) + \
    geom_boxplot(aes('y', 'x'), orientation='y')
```
