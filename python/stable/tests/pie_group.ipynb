```python
import geopandas as gpd
import shapely as sh

from lets_plot import *
LetsPlot.setup_html()
```

```python

data = {
    "City_Code": [ "A", "A", "B", "B" ], 
    "City_Name": [ "City A", "City A", "City B", "City B" ],
    "Vote": [ "Yes", "No", "Yes", "No" ], 
    "Number": [ 120.0, 30.0, 20.0, 80.0 ], 
}

cities = gpd.GeoDataFrame({
    "id": [ "A", "B" ],
    "geometry": [
        sh.geometry.Point(-80.0, -40.0),
        sh.geometry.Point(80.0, 40.0)
    ]
}, crs="EPSG:4326")

ggplot(data) \
    + geom_pie(
        aes(fill="Vote", weight="Number", group="City_Code"), 
        map=cities,
        map_join=["City_Code", "id"], 
    ) \
    + scale_x_continuous(expand=[0.3]) \
    + scale_y_continuous(expand=[0.3])
```
