scale_y_log10: hidden segments

```python
n = 10
x = list(range(n))
y = [2**d for d in range(n)]
p = ggplot({'x': x, 'y': y}) + geom_point(aes('x', 'y')) + ggsize(300, 200)

plots = GGBunch()
plots.add_plot(p, 0, 0)
plots.add_plot(p + scale_y_log10(), 300, 0)
plots.add_plot(p + geom_segment(aes(x='x', y=[0]*n, xend='x', yend='y')), 0, 200)
plots.add_plot(p + geom_segment(aes(x='x', y=[0]*n, xend='x', yend='y')) + scale_y_log10(), 300, 200)
plots.show()
```
