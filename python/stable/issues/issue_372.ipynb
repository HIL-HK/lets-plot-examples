`scale_y_log10`: hidden segments

Code here will produce plot without segments, but
the workaround would be:
a) define segment 'y' as constant: geom_segment(aes(x='x', xend='x', yend='y'), y=0)
b) make 'y'-s not 0 but just small: geom_segment(aes(x='x', y=[1e-3]*n, xend='x', yend='y'))

```python
n = 10
x = list(range(n))
y = [2**d for d in range(n)]
p = ggplot({'x': x, 'y': y}) + geom_point(aes('x', 'y')) + ggsize(300, 200)

gggrid([p, p + scale_y_log10(), 
        p + geom_segment(aes(x='x', y=[0]*n, xend='x', yend='y')), 
        p + geom_segment(aes(x='x', y=[0]*n, xend='x', yend='y')) + scale_y_log10()], 
      ncol=2)
```

```python
p + geom_segment(aes(x='x', y=[0]*n, xend='x', yend='y'))
```

```python
p + geom_segment(aes(x='x', y=[0]*n, xend='x', yend='y')) + scale_y_log10()
```
