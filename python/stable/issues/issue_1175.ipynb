
`sampling_vertex_vw` doesn't work as expected 

(The quality of the resulting geometries is unacceptably bad)

https://github.com/JetBrains/lets-plot/issues/1175 [closed]

```python
from lets_plot import *
import numpy as np

LetsPlot.setup_html()
```

```python
def f(x, y, n):
    z = np.sin(x * 3 * np.pi / n)
    z += np.sin(y * 3 * np.pi / n)
    z += x * 3 / n
    z += y * 5 / n
    return z

n = 50
x = np.arange(n)
y = np.arange(n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y, n)

data = dict(
    x=X.reshape(-1),
    y=Y.reshape(-1),
    z=Z.reshape(-1))

```

```python
p = ggplot(data, aes('x','y', z='z')) \
    + theme_void() + ggsize(200, 200)
```

```python
gggrid([
    p + geom_contour() + ggtitle("No sampling"),
    p + geom_contour(sampling=sampling_vertex_vw(100)) + ggtitle("sampling_vertex_vw(100)"),
    p + geom_contour(sampling=sampling_vertex_dp(100)) + ggtitle("sampling_vertex_dp(100)")
])  
```

```python

p = ggplot(data, aes('x','y', z='z', fill='..level..', color='..group..')) \
    + theme_void() + ggsize(300, 300) + theme(legend_position='none')

gggrid([
    p + geom_contourf(bins=6, sampling=sampling_vertex_vw(n=8000)),
    p + geom_contourf(bins=6, sampling=sampling_vertex_vw(n=100))
])      
```
