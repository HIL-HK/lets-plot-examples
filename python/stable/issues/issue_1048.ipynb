Livemap: parameters nudge_x and nudge_y have no effect on geom_text or geom_label

https://github.com/JetBrains/lets-plot/issues/1048

```python
import numpy as np
from lets_plot import *
from lets_plot.geo_data import *

LetsPlot.setup_html()
```

```python

france = geocode_countries(['France'])
germany = geocode_countries(['Germany'])
```

```python
gggrid([ggplot() + geom_livemap() + geom_label(aes(label='found name'), nudge_y=1, map=france, color='red', size=10) + \
        geom_text(aes(label='found name'), nudge_y=1, map=germany, color='#02633e', fontface='bold') + ggtitle("With nudge_y"),
        ggplot() + geom_livemap() + geom_label(aes(label='found name'), map=france, color='red', size=10) + \
        geom_text(aes(label='found name'), map=germany, color='#02633e', fontface='bold')]) + \
    theme(plot_title=element_text(family='bold', size=20, hjust=0.5, color='blue')) + ggtitle('Grid Title')
```
