`element_blank()` has no effect on plot title/subtitle/caption in theme() [CLOSED]

" ... +  theme(plot_title=element_blank(), plot_subtitle=element_blank(), plot_caption=element_blank())"

does not hide these elements on the plot.

https://github.com/JetBrains/lets-plot/issues/913

```python
mpg = pd.read_csv ("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg
```

```python
p = ggplot(mpg) + geom_point(aes('displ', 'hwy', color = 'drv'))

p1 = p + ggtitle("The plot title using 'ggtitle()'")

p1
```

```python
p1 + theme(plot_title=element_blank())
```

```python
p2 = p + labs(title = "The plot title", subtitle = "The plot subtitle", caption = "The plot caption")

p2
```

```python
p2 + theme(plot_title=element_blank(), plot_subtitle=element_blank(), plot_caption=element_blank())
```
