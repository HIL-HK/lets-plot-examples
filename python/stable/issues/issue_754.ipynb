`geom_boxplot()` doesn't apply alpha to outliers [CLOSED]

https://github.com/JetBrains/lets-plot/issues/754

"Outliers appearance is also controlled by additional parameters outlier_[color,fill,shape,size].
In this vein, we could add two more parameters: outlier_alpha, outlier_stroke."

```python
n = 100
np.random.seed(42)
x = np.random.choice(['a', 'b', 'c'], size=n)
y = np.random.normal(size=n)
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + \
    geom_boxplot(alpha=0.2)
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + \
    geom_boxplot(color='red', fill='light_blue', outlier_alpha=0.4, size=2)
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + \
    geom_boxplot(color='red', fill='light_blue', alpha=0.4, size=2)
```
