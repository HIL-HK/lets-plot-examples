The values of vertical text alignment (vjust parameter) are between 0 and 1. But 1 (maximum shift) is sometimes too small to separate points and texts.

We need to support nudge_x, nudge_y parameters (#74, #324).

```python

data = {'x': [0], 'y': [0], 't': ['Point']}

ggplot(data, aes(x='x', y='y')) + \
    geom_point() + geom_text(aes(label='t'), vjust=1)
```

```python
ggplot(data, aes(x='x', y='y')) + \
    geom_point() + geom_text(aes(label='t'), vjust=1, hjust='inward')
```

```python
ggplot(data, aes(x='x', y='y')) + \
    geom_point() + geom_label(aes(label='t'), hjust='inward')
```

```python
ggplot(data, aes(x='x', y='y')) + geom_point() + \
    geom_text(aes(label='t'), hjust=1, vjust=0, lineheight=0.7)
```

```python
ggplot(data, aes(x='x', y='y')) + geom_point() + \
    geom_text(aes(label='t'), nudge_y=0.02)
```
