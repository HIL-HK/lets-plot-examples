Can't set None for coord limit

```python
p = ggplot() + geom_line(
    aes(x='x', y='y', group='g', color='g', size='g'), data=
    {
    'x': [0, 5, 10, 15, 20, 25],
    'y': [0, 5, 10, 15, 20, 25],
    'g': ['a', 'a', 'b', 'b', 'c', 'c']
}) + theme(legend_position='none') + ggsize(300, 300)

b = GGBunch()
b.add_plot(p + coord_fixed() + ggtitle("coord_fixed()"), 0, 0)
b.add_plot(p + coord_fixed(xlim=(12, 22)) + ggtitle("coord_fixed(xlim=(12, \n22))"), 300, 0)
b.add_plot(p + coord_fixed(xlim=(None, 22)) + ggtitle("coord_fixed(xlim=(None, 22))"), 600, 0)
b.show()
```
