Wrong tooltip position on `geom_segment()` with position adjustment [CLOSED]

https://github.com/JetBrains/lets-plot/issues/963

```python

data = {
    'x': [0, 1],
    'y': [1, 1],
    'xend': [1, 2],
    'yend': [2, 2],
}

ggplot(data) + \
    geom_segment(aes('x', 'y', xend='xend', yend='yend'), tooltips=layer_tooltips().line('Tooltip'))
```

```python
ggplot(data) + \
    geom_segment(aes('x', 'y', xend='xend', yend='yend'), 
                 position=position_nudge(x=.5),
                 tooltips=layer_tooltips().line('Tooltip'))
```

```python
ggplot(data) + \
    geom_segment(aes('x', 'y', xend='xend', yend='yend'), 
                 position=position_jitter(width=.5, height=0, seed=42),
                 tooltips=layer_tooltips().line('Tooltip'))
```
