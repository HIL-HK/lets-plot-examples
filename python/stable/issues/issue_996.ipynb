`geom_spoke()`: axis tooltips don't appear

https://github.com/JetBrains/lets-plot/issues/996

The axis tooltips will appear if you swap the point and spoke layers in this example.

```python
data = {
    'x': [0, 1],
    'y': [0, 1],
    'a': [0.79, 3.93],
    'r': [.71, .35],
}

ggplot(data, aes('x', 'y')) + \
    geom_point() + \
    geom_spoke(aes(angle='a', radius='r')) + \
    coord_fixed()
```

```python
ggplot(data, aes('x', 'y')) + \
    geom_spoke(aes(angle='a', radius='r')) + \
    geom_point() + \
    coord_fixed()
```
