geom_livemap: can't draw a path crossing the antimeridian:

https://github.com/JetBrains/lets-plot/issues/649

```python
from lets_plot import *
LetsPlot.setup_html()

short = {
    'x': [-170, 170],
    'y': [0, 0]
}

long = {
    'x': [-170, 0, 170],
    'y': [0, 0, 0]
}

p = ggplot() \
    + geom_livemap(zoom=1, location=[180, 0]) \
    + geom_point(x=-170, y=0, color='red') \
    + geom_point(x=170, y=0, color='blue')

gggrid([
    p + geom_path(aes('x', 'y'), data=short) + ggtitle("'x': [-170, 170]"),
    p + geom_path(aes('x', 'y'), data=long) + ggtitle("'x': [-170, 0, 170]"),
], ncol=1)
```
