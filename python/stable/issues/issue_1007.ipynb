```python
import pandas as pd
import lets_plot as lp
from lets_plot.mapping import as_discrete
```

how to align ticks with ggrid?

https://github.com/JetBrains/lets-plot/issues/1007

```python

lp.LetsPlot.setup_html()

df_female = pd.DataFrame(data={'Industry': ['Construction', 'Mining', 'National',
                                                 'Construction', 'Mining', 'National',
                                                 'Construction', 'Mining', 'National'],
                                                 'Category': ['Female', 'Female', 'Female',
                                                 'Female Manager', 'Female Manager', 'Female Manager',
                                                 'Diff', 'Diff', 'Diff'],
                                    'Proportion': [26.0,  22.0, 51.2, 18.9, 22.9, 42.0, 7.1, 0.9, 9.2]})
theme_blank_xaxis = lp.theme(axis_title_x=lp.element_blank(),
                             axis_text_x=lp.element_blank(),
                             axis_ticks_x=lp.element_blank())

ordered_industry = as_discrete('Industry', order=-1)
p3a = (lp.ggplot(df_female[df_female.Category == 'Female'], lp.aes(weight='Proportion',
                                                                             x=ordered_industry, fill='Category')) +
    lp.geom_bar(lp.aes(fill='Category'),
            labels=lp.layer_labels().line('@..count..').format(field='@..count..',format=".0f").size(15), alpha=0.7) +
    lp.labs(title=' Proportion by industry, 2023)',
              x='', y='Female Proportion (%)', fill='') +
    lp.scale_fill_manual(values=['#FFC300'], name='',
                           breaks=['Female_p'], labels=['% Women']) +
    #lp.scale_x_discrete(expand=[0.0,0.2]) +
    lp.theme_bw() + lp.theme(legend_position="top") + theme_blank_xaxis)

p3b = (lp.ggplot(df_female[df_female.Category == 'Diff'], lp.aes(y='Proportion',
                                                                             x=ordered_industry)) +
#       lp.scale_x_discrete(expand=[0.05,0.8]) +
    lp.geom_point() + lp.geom_line() + lp.theme_bw())

p3c = lp.gggrid([p3a,p3b], ncol = 1, sharex=True, heights=[3,2]) 

p3c.show()
```

First of all you'll need to add parameter `align=True` to `gggrid()`

However, at the moment `sharex` option doesn't work with discrete scales. 

(`gggrid`: sharex, sharey options do not work with discrete scales. 
https://github.com/JetBrains/lets-plot/issues/1015)

```python
p3d = lp.gggrid([p3a,p3b], ncol = 1, sharex=True, heights=[3,2], align=True) 

p3d.show()
```

But, you can still align x-scales in your example by adding expands to the line-plot's x-scale:

`... + scale_x_discrete(expand=[0.0, 0.2 + 0.95/2])`

```python
p3a = (lp.ggplot(df_female[df_female.Category == 'Female'], lp.aes(weight='Proportion',
                                                                             x=ordered_industry, fill='Category')) +
    lp.geom_bar(lp.aes(fill='Category'),
            labels=lp.layer_labels().line('@..count..').format(field='@..count..',format=".0f").size(15), alpha=0.7) +
    lp.labs(title='Proportion by industry, 2023',
              x='', y='Female Proportion (%)', fill='') +
    lp.scale_fill_manual(values=['#FFC300'], name='',
                           breaks=['Female_p'], labels=['% Women']) +
    lp.theme_bw() + lp.theme(legend_position="top") + theme_blank_xaxis)

p3b = (lp.ggplot(df_female[df_female.Category == 'Diff'], lp.aes(y='Proportion',
                                                                             x=ordered_industry)) +
       lp.geom_point() + lp.geom_line() + lp.theme_bw() +
       lp.scale_x_discrete(expand=[0.0, 0.2 + 0.95/2]))

p3c = lp.gggrid([p3a,p3b], ncol = 1, sharex=True, heights=[3,2], align=True) 

p3c.show()
```
