Unexpected replacement of double curly brackets with a single curly bracket. 
[#1433](https://github.com/JetBrains/lets-plot/issues/1433) [closed]


```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
df = {'formula': ["\\( x^{a^{b}}+1\\)"]}

p = ggplot(df, aes(color='formula')) \
    + geom_point(tooltips=layer_tooltips().line("@formula")) \
    + theme(text=element_text(size=20))
p
```

```python
# Adding a space solves the issue:

df1 = {'formula': ["\\( x^{a^{b} }+1\\)"]}

(ggplot(df1, aes(color='formula')) 
    + geom_point(tooltips=layer_tooltips().line("@formula")) 
    + theme(text=element_text(size=20)))
```

```python
# With regular text:

df2 = {'text': ["{{foo}}"]}

(ggplot(df2, aes(color='text')) 
    + geom_point(tooltips=layer_tooltips().line("@text")) 
    + theme(text=element_text(size=20)))
```
