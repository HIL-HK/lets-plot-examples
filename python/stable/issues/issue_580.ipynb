```python
import numpy as np
from lets_plot import *
LetsPlot.setup_html()
```

The tooltip not rotate after we apply `coord_flip()`.

https://github.com/JetBrains/lets-plot/issues/580

```python
n = 200
np.random.seed(42)

x = np.repeat(['A','B'], n)
y = np.concatenate((np.random.normal(0, 1, n), np.random.normal(.8, 1, n)))

ggplot({'x': x, 'y': y}, aes(x='x', y='y', color='x')) + geom_boxplot()
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y', color='x')) + geom_boxplot() + coord_flip()
```
