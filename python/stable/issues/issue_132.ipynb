https://github.com/JetBrains/lets-plot/issues/132 [closed]


```python
data = {
    'label': ["==    0 ==>", "==   60 ==>", "==  120 ==>", "==  180 ==>", "== -60 ==>", "== -120 ==>"],
    'angle': [0, 60, 120, 180, -60, -120],
}
```

```python
ggplot(data) + geom_livemap() + geom_point(x=0, y=0, size=140, color='white', alpha=0.6)\
  + geom_text(aes(label = 'label', angle='angle'), data=data,
              x=0, y=0, 
              hjust = "left", vjust = "center",
              size=10, family='monospace', color='red')
```

```python
ggplot(data) + geom_text(aes(label = 'label', angle='angle'), 
                         data=data, x=0, y=0, 
                         hjust = "left", vjust = "center",
                         size=10, family='helvetica', color='red')
```
