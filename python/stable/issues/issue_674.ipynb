Area ridges: 'viridis twilight' fill overlaps geometry borders [CLOSED]

https://github.com/JetBrains/lets-plot/issues/674


```python
data = {
    "x": [0, 0, 1, 1, 2, 2, 3, 3, 4],
    "y": [0, 0, 0, 0, 0, 0, 0, 0, 0],
    "h": [1, 1, .8, .8, .2, .2, .8, .8, 1],
    "q": [0, 0, 0, .5, .5, .8, .8, 1, 1],
}
```

```python
ggplot(data, aes("x", "y")) + \
    geom_area_ridges(aes(height="h", quantile="q", fill="q"), stat='identity') + \
    scale_fill_viridis(option='twilight')
```

```python
ggplot(data, aes("x", "y")) + \
    geom_area_ridges(aes(height="h", quantile="q", fill="q"), stat='identity') + \
    scale_fill_gradientn(colors=["red", "blue", "red"])
```

```python
ggplot(data, aes("x", "y")) + \
    geom_area_ridges(aes(height="h", quantile="q", fill="q"), stat='identity') + \
    scale_fill_viridis(option='magma')
```

```python
ggplot(data, aes("x", "y")) + \
    geom_area_ridges(aes(height="h", quantile="q", fill="q"), stat='identity') + coord_flip() + \
    scale_fill_viridis(option='magma')
```

```python
ggplot(data) + \
    geom_area_ridges(aes("x", "y", height="h", quantile="q", fill="q"), stat='identity') + \
    scale_fill_gradientn(colors=["blue", "red", "blue"])
```

```python
ggplot(data) + \
    geom_area(aes("x", "h", quantile="q", fill="q"), stat='identity', color="black" ) + \
    scale_fill_gradientn(colors=["blue", "red", "blue"])
```

```python
ggplot(data) + \
    geom_area(aes("h", "x", quantile="q", fill="q"), stat='identity', color="black", orientation='y' ) + \
    scale_fill_gradientn(colors=["blue", "red", "blue"])
```

```python
ggplot(data) + \
    geom_area(aes("x", "h", quantile="q", fill="q"), stat='identity', color="black")  + coord_flip()+ \
    scale_fill_gradientn(colors=["blue", "red", "blue"])
```

```python
ggplot(data) + \
    geom_violin(aes("y", "x", violinwidth="h", quantile="q", fill="q"), stat='identity', show_half=1) + \
    scale_fill_gradientn(colors=["blue", "red", "blue"])
```

```python
ggplot(data) + \
    geom_violin(aes("y", "x", violinwidth="h", quantile="q", fill="q"), stat='identity', show_half=1) + coord_flip() + \
    scale_fill_gradientn(colors=["blue", "red", "blue"])
```
