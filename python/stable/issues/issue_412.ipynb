https://github.com/JetBrains/lets-plot/issues/412 [Closed]

`geom_jitter`: show axis tooltips (same as geom_point)

```python
df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv')
```

```python
ggplot(df, aes('cty', 'hwy')) + geom_point()
```

```python
ggplot(df, aes('cty', 'hwy')) + geom_jitter()
```

```python
ggplot(df, aes('fl', 'drv')) + geom_jitter(width=.3, height=.3, )
```
