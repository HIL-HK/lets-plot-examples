# [Incorrect appearance of stacked density plot in polar coordinates](https://github.com/JetBrains/lets-plot/issues/1375)

```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'x': [1, 1, 1, 2],
    'g': ['a', 'a', 'b', 'b']
}
ggplot(data) + \
    geom_density(aes(x='x', fill='g'), position='stack') + \
    coord_polar()
```
