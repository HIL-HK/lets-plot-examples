Add horizontal segment for boxplot whiskers and option to control its size

https://github.com/JetBrains/lets-plot/issues/549

```python
n = 100
np.random.seed(42)
x = np.random.choice(['a', 'b', 'c'], size=n)
y = np.random.normal(size=n)

ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + geom_boxplot(whisker_width=0.2)
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + geom_boxplot()
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + geom_boxplot(whisker_width=1)
```

```python
ggplot({'x': x, 'y': y}, aes(x='x', y='y')) + geom_boxplot(whisker_width=1) + coord_flip()
```
