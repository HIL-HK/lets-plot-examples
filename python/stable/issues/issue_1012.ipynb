No gridlines when axis_ontop=True [CLOSED]

https://github.com/JetBrains/lets-plot/issues/1012

```python
import pandas as pd
from lets_plot import *

LetsPlot.setup_html()

plot_data = pd.DataFrame.from_records([
    ("pet", "cat", 5, "carnivore"),
    ("pet", "dog", 10, "carnivore"),
    ("pet", "rabbit", 2, "herbivore"),
    ("pet", "hamster", 1, "herbivore"),

    ("farm_animal", "cow", 500, "herbivore"),
    ("farm_animal", "pig", 100, "carnivore"),
    ("farm_animal", "horse", 700, "herbivore"),
])
plot_data.columns = ("animal_type", "animal", "weight", "diet")


plot = (
    ggplot(plot_data, aes(x="animal", y="weight"))
    + geom_bar(stat="identity")
    + theme_bw()
    + theme(
        axis_ontop=True
    )
)
plot
```

```python
plot = (
    ggplot(plot_data, aes(x="animal", y="weight"))
    + geom_bar(stat="identity")
    + theme_bw()
    + theme(
        axis_ontop=False
    )
)
plot
```

```python

plot = (
    ggplot(plot_data, aes(x="animal", y="weight"))
    + geom_bar(stat="identity", size = 3, color = 'red')
    + theme_bw()
    + theme(
        axis_ontop=True
    )
)
plot
```

```python

plot = (
    ggplot(plot_data, aes(x="animal", y="weight"))
    + geom_bar(stat="identity", size = 3, color = 'red')
    + theme_bw()
    + theme(
        axis_ontop=False
    )
)
plot
```
