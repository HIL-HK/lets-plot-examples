`geom_bin2d`: wrong geometry size when `bins=[n, 1]`

Expected:



```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

## Original Examples

```python
data = {
    'x': [0, 0, 1, 2],
    'y': [0, 0, 1, 2],
}
```

```python
ggplot(data, aes('x', 'y')) + geom_bin2d(bins=[2, 1]) + geom_point(color="red")
```

```python
ggplot(data, aes('x', 'y')) + geom_bin2d(bins=[1, 2]) + geom_point(color="red")
```

```python
ggplot(data, aes('x', 'y')) + geom_bin2d(bins=[1, 1]) + geom_point(color="red")
```

## Additional Tests

```python
def get_test(data, bins, geom):
    return ggplot(data, aes('x', 'y')) + \
        geom(bins=bins, color="white", size=.5) + \
        geom_point(color="red") + \
        ggtitle("{geom}(bins={bins})".format(geom=geom.__name__, bins=bins))
```

```python
data1 = {
    'x': [0, 0, 1, 2],
    'y': [0, 0, 1, 2],
}

data2 = {
    'x': [0, 0, 20, 40],
    'y': [0, 0, 0, 0],
}

data3 = {
    'x': [0, 0, 0, 0],
    'y': [0, 0, 20, 40],
}
```

```python
gggrid([
    get_test(data1, [2, 1], geom_bin2d),
    get_test(data1, [1, 2], geom_bin2d),
    get_test(data1, [1, 1], geom_bin2d),
    get_test(data1, [2, 1], geom_hex),
    get_test(data1, [1, 2], geom_hex),
    get_test(data1, [1, 1], geom_hex),

    get_test(data2, [2, 1], geom_bin2d),
    get_test(data2, [1, 2], geom_bin2d),
    get_test(data2, [1, 1], geom_bin2d),
    get_test(data2, [2, 1], geom_hex),
    get_test(data2, [1, 2], geom_hex),
    get_test(data2, [1, 1], geom_hex),

    get_test(data3, [2, 1], geom_bin2d),
    get_test(data3, [1, 2], geom_bin2d),
    get_test(data3, [1, 1], geom_bin2d),
    get_test(data3, [2, 1], geom_hex),
    get_test(data3, [1, 2], geom_hex),
    get_test(data3, [1, 1], geom_hex),
], ncol=3)
```
