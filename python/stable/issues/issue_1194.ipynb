Incorrect Y-axis layout with `facets` and `panel_inset`

https://github.com/JetBrains/lets-plot/issues/1194 [closed]


```python
d = {
    l: [i]*2 for i, l in enumerate(string.ascii_lowercase)
    
} | {'facet': [0, 1]}

ggplot(d, aes(x='a', y='b')) + geom_point(tooltips=layer_tooltips(list(string.ascii_lowercase))) \
    + theme(panel_inset=[400, 0, 0]) + ggsize(800, 1000) + facet_wrap("facet")
```

```python
# without panel_inset

ggplot(d, aes(x='a', y='b')) + geom_point(tooltips=layer_tooltips(list(string.ascii_lowercase))) \
    + ggsize(800, 1000) + facet_wrap("facet")
```
