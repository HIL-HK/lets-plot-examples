https://github.com/JetBrains/lets-plot/issues/608 [closed]

`element_blank()` has no effect in theme `legend_title`

+ theme(legend_title=element_blank())

Does not suppress the legend name in the plot.
It happens in both a facet_grid or a single bar plot. The fill column name which defines the legend title is always displayed.

```python
mpg_df=pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv')
```

```python
ggplot(mpg_df) + \
    geom_bar(aes(x='cyl', fill='drv'), color='white', size=0.5)
```

```python
ggplot(mpg_df) + \
    geom_bar(aes(x='cyl', fill='drv'), color='white', size=0.5) + \
    theme(legend_title=element_blank())
```
