Wide axis labels should not be cropped

https://github.com/JetBrains/lets-plot/issues/906

```python

n = 5
x = [i * 10**(-5) for i in range(n)]
y = [0] * len(x)

ggplot(dict(x=x, y=y), aes('x', 'y')) + \
    geom_point() + \
    scale_x_continuous(format='.7e')
```

```python
ggplot(dict(x=x, y=y), aes('x', 'y')) + \
    geom_point() + theme(exponent_format='pow') + scale_x_continuous() 
```

```python
ggplot(dict(x=x, y=y), aes('x', 'y')) + \
    geom_point() + theme(exponent_format='pow') 
```
