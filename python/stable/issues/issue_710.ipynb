```python
import numpy as np
import pandas as pd
from lets_plot import *

LetsPlot.setup_html()
```

Option to save plots to PDF format using `ggsave()`

https://github.com/JetBrains/lets-plot/issues/710 [closed]

```python

data = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg2.csv')

p = (ggplot(data, aes(x="engine horsepower", y="miles per gallon")) + \
     geom_point(aes(color="origin of car")) + \
     ggsize(800, 350))

p
```

```python
pathname = ggsave(p, "plot.svg")
```

```python
from IPython.display import SVG

display(SVG(filename=pathname))
```

```python
import cairosvg

cairosvg.svg2pdf(url=pathname, write_to='lets-plot-images/plot.pdf')

# Check the file plot.pdf in "lets-plot-images" folder 
```
