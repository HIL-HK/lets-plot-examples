When we use something like `coord_fixed(...)`, the cropped polygons lose their tooltips.

```python
x = [-1, 1, 1, -1, 1, 3, 3, 1]
y = [-1, -1, 1, 1, -1, -1, 1, 1]
g = ['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b']

p = ggplot({'x': x, 'y': y, 'g': g}, aes('x', 'y')) + \
    geom_polygon(aes(group='g', fill='g'), \
                 tooltips=layer_tooltips().line('@g'))

bunch = GGBunch()
bunch.add_plot(p, 0, 0, 400, 200)
bunch.add_plot(p + coord_fixed(xlim=[-2, 2]), 400, 0, 400, 200)
bunch.show()
```
