## Test example

```python
expected_data = {
    'x': [0, 1, 2],
    'y': [2, 1, 3],
}
actual_data = {
    'x': [0, 2, 1],
    'y': [2, 3, 1],
}
```

```python
gggrid([
    ggplot(expected_data, aes('x', 'y')) + geom_area() + ggtitle("Expected"),
    ggplot(actual_data, aes('x', 'y')) + geom_area() + ggtitle("Actual"),
])
```
