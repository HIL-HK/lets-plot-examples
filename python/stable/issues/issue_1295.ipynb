# [Issue #1295](https://github.com/JetBrains/lets-plot/issues/1295) - Incorrectly rendered Area chart

```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

## Test example

```python
expected_data = {
    'x': [0, 1, 2],
    'y': [2, 1, 3],
}
actual_data = {
    'x': [0, 2, 1],
    'y': [2, 3, 1],
}
```

```python
gggrid([
    ggplot(expected_data, aes('x', 'y')) + geom_area() + ggtitle("Expected"),
    ggplot(actual_data, aes('x', 'y')) + geom_area() + ggtitle("Actual"),
])
```

## User example

```python
df = pd.read_csv("https://github.com/user-attachments/files/18707097/generated_jobseeker_test_data.csv")
```

```python
gggrid([
    ggplot(df.sort_values(by="kpi_month_la"), aes(x='kpi_month_la', y='jobseekers', fill='mau_type')) + geom_area() + ggtitle("Expected"),
    ggplot(df, aes(x='kpi_month_la', y='jobseekers', fill='mau_type')) + geom_area() + ggtitle("Actual")
])
```
