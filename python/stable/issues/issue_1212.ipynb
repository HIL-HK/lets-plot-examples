Error when using scale_identity(aesthetic="shape") [closed]

https://github.com/JetBrains/lets-plot/issues/1212

`+ scale_shape_identity()` works as expected but
`+ scale_identity(aesthetic="shape")` produces error:

Internal error: IllegalStateException : [IdentityDiscreteMapperProvider] Can't create mapper for continuous domain DoubleSpan(21, 22)

The same error when using scale_identity(aesthetic="linetype")

```python
import numpy as np
from lets_plot import *
LetsPlot.setup_html()
```

```python
df = {'cond': [3, 3, 4, 4, 5, 5],
      'x': [2, 3, 2, 3, 2, 3],
      'y': [1.0, 1.5, 2.0, 1.0, 1.6, 1.8]}

ggplot(df) + \
    geom_point(aes(x='x', y='y', shape='cond'), size=7) + \
    scale_shape_identity()
```

```python
ggplot(df) + \
    geom_point(aes(x='x', y='y', shape='cond'), size=7) + \
    scale_identity(aesthetic='shape')
```

```python
ggplot(df) + geom_line(aes(x='x', y='y', group='cond', linetype='cond')) + \
    geom_point(aes(x='x', y='y', shape='cond'), size=7) + \
    scale_linetype_identity() + \
    scale_shape_identity()
```

```python
ggplot(df) + geom_line(aes(x='x', y='y', group='cond', linetype='cond')) + \
    geom_point(aes(x='x', y='y', shape='cond'), size=7) + \
    scale_identity(aesthetic=['shape', 'linetype'])
```
