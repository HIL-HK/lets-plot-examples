`ggsave()` doesn't save `geom_raster()` layer to a file 

https://github.com/JetBrains/lets-plot/issues/778

```python
data = {
    "x": [ -1.0, 1.0, -1.0, 1.0 ],
    "y": [ -1.0, -1.0, 1.0, 1.0 ],
    "z": [ 0.024, 0.094, 0.094, 0.024 ]
}
p = ggplot(data) + \
    geom_raster(aes(x='x', y='y', fill='z'), data=data) + \
    geom_point(aes(x='x', y='y'), size=8, color='red', data=data) + \
    scale_fill_gradient(low="#54278f", high="#f2f0f7")

print("Plot")
p.show()

print("Image")
plot_filename = ggsave(p, 'issue_ggsave_raster1.svg')
display(SVG(plot_filename))
```

```python
p1 = ggplot(data) + \
    geom_raster(aes(x='x', y='y', fill='z'), data=data)
p1.show()

print("Image")
ggsave(p1, 'plot1.png')
```
