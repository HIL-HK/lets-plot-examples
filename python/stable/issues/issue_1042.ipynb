Assigning same alpha value to all datapoints causes unnecessary transparency [CLOSED]

https://github.com/JetBrains/lets-plot/issues/1042

```python
import numpy as np
from lets_plot import *

LetsPlot.setup_html()
```

```python
x = np.arange(1, 10)
y = x[::-1]
z = [1] * len(x)
data = {'x': x, 'y': y, 'z': z}

(ggplot(data, aes(x='x', y='y', alpha='z'))
    + geom_bar(stat='identity', fill='#2b8cbe', tooltips='none')
    + guides(alpha='none') 
    + geom_label(aes(label='y'), nudge_y=1, label_format='.1f', fill='#2b8cbe')
    + scale_alpha_identity() 
)
```

```python
(ggplot(data, aes(x='x', y='y', alpha='z'))
    + geom_bar(stat='identity', fill='#2b8cbe', tooltips='none')
    + guides(alpha='none') 
    + geom_label(aes(label='y'), nudge_y=1, label_format='.1f', fill='#2b8cbe')
)
```

```python
(ggplot(data, aes(x='x', y='y'))
    + geom_bar(stat='identity', fill='#2b8cbe', tooltips='none', alpha=0.2, )
    + guides(alpha='none') 
    + geom_label(aes(label='y'), nudge_y=1, label_format='.1f', fill='#2b8cbe',alpha=0.2, alpha_stroke=0.2)
)
```
