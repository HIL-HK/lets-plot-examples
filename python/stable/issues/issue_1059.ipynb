Livemap: Add zooming-in for `geom_text()`/`geom_label()` 

https://github.com/JetBrains/lets-plot/issues/1059

```python
import pandas as pd
from lets_plot import *
from lets_plot.geo_data import *

LetsPlot.setup_html()
```

```python
france = geocode_countries(['France'])
germany = geocode_countries(['Germany'])
```

```python
ggplot() + geom_livemap() + geom_label(aes(label='found name'), nudge_y=1, map=france, color='red', size=10) + \
        geom_text(aes(label='found name'), nudge_y=1, map=germany, color='#02633e', fontface='bold')
```

```python
# No scale - const_size_zoomin=0

(ggplot() + geom_livemap(const_size_zoomin=0) 
    + geom_label(aes(label='found name'), nudge_y=1, map=france, color='red', size=10) 
    + geom_text(aes(label='found name'), nudge_y=1, map=germany, color='#02633e', fontface='bold'))
```

```python
# One step scale

(ggplot() + geom_livemap(const_size_zoomin=1) 
    + geom_label(aes(label='found name'), nudge_y=1, map=france, color='red') 
    + geom_text(aes(label='found name'), nudge_y=1, map=germany, color='#02633e'))
```
