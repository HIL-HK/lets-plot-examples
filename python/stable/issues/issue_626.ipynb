`ggsave` (.svg) transforms `geom_text` integer to float

https://github.com/JetBrains/lets-plot/issues/626 [closed]

```python
from lets_plot import *
from IPython.display import SVG, Image, HTML
LetsPlot.setup_html()


df = {
    'x': [0],
    'y': [0],
    'label': [777]
}

p = ggplot(df) + geom_label(aes(x='x', y='y', label='label'))
p
```

```python
display(SVG(ggsave(p, 'issue_626.svg')))
```

```python
from datetime import datetime

df = {
    'x': [0],
    'y': [0],
    'label': [777],
    'daytime': [datetime(2024, 1, 1, 12, 34, 56)]
}

p = ggplot(df) + geom_label(aes(x='x', y='y', label='daytime'), tooltips=layer_tooltips().line('^x, ^y: @daytime'))
p
```

```python
display(SVG(ggsave(p, 'issue_626.svg')))
```
