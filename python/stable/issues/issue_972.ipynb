Allow `geom_bar()` as a marginal layer

https://github.com/JetBrains/lets-plot/issues/972

```python

data = {
    'x': ["a", "a", "b", "b"],
    'y': ["f", "g", "g", "g"],
}

ggplot(data, aes('x', 'y')) + \
    geom_count() + \
    ggmarginal('r', layer=geom_bar(orientation='y'))
```
