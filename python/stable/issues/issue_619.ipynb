Positional "constant" doesn't honor axis limits

```python
# No limits - both labels are visible

p = (ggplot() +
    geom_label(aes(y=[10]), label="Data y=10") +
    geom_label(label="Const y=13", y = 13)
)
p
```

```python
# y-axis is limited : "Const y=13" should not be visible.
p + ylim(10, 10)
```
