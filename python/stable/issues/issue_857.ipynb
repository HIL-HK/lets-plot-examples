`theme / flavor` plot background is not shown on plot with `geom_livemap()` layer. [CLOSED]

Link: https://github.com/JetBrains/lets-plot/issues/857

```python

d = {
    'x': [20, 30, 40],
    'y': [60, 60, 60],
    's': [1, 5, 10],
    'g': ['A', 'B', 'C']
}

p = ggplot(d, aes(x='x', y='y', color='g', size='s')) + flavor_darcula()
gggrid([
    p + geom_point() + ggtitle("geom_point()"),
    p + geom_livemap() + geom_point() + ggtitle("geom_livemap() + geom_point()")
])
```

```python
# без рамок в легенде у livemap

p = ggplot(d, aes(x='x', y='y')) + flavor_darcula()
gggrid([
    p + geom_point(aes(color='g', size='s')) + ggtitle("geom_point()"),
    p + geom_livemap() + geom_point(aes(color='g', size='s')) + ggtitle("geom_livemap() + geom_point()")
])
```

```python

# фасеты и параметры фона

p + geom_livemap() + geom_point(aes(color='g', size='s')) + facet_wrap(facets='g', ncol=3) \
    + theme(plot_background=element_rect(color='orange', size=10)) + ggsize(800, 300)
```
