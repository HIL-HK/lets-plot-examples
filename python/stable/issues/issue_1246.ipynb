Badly formatted zero break for the "~g" format [CLOSED]

https://github.com/JetBrains/lets-plot/issues/1246

The same applies to the "\~e" and "\~s" formats.

```python
data = {
    "x": [1.7, 1.4, 1.5, 1.4, 1.5],
    "y": [
      0.05999999999999994,
      0.10999999999999982,
      -0.04000000000000009,
      0.009999999999999842,
      -0.1400000000000001
    ]
}

ggplot(data, aes('x', 'y')) + geom_point() + scale_y_continuous(format="~g")
```

```python
ggplot(data, aes('x', 'y')) + geom_point() + scale_y_continuous(format="g")
```

```python
ggplot(data, aes('x', 'y')) + geom_point() + scale_y_continuous(format="~s")
```

```python
ggplot(data, aes('x', 'y')) + geom_point() + scale_y_continuous(format="~e")
```

```python
ggplot(data, aes('x', 'y')) + geom_point() + scale_y_continuous(format="e")
```
