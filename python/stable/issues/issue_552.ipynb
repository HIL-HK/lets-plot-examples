```python
import numpy as np
from lets_plot import *

LetsPlot.setup_html()
```

Currently, lines are drawn on top of all filled areas. 
Expected: In the example, the group 'A' line should be partially obscured by the group 'B' area.

https://github.com/JetBrains/lets-plot/issues/552

```python
np.random.seed(12)
data = dict(
    cond=np.repeat(['A','B'], 200),
    rating=np.concatenate((np.random.normal(0, 1, 200), np.random.normal(1, 1.5, 200)))
)
```

```python
ggplot(data, aes(x='rating', fill='cond')) + ggsize(500, 250) \
+ geom_density(color='black', size=4, alpha=1) + scale_fill_brewer(type='seq')
```

```python
ggplot(data, aes(x='rating', fill='cond')) + ggsize(500, 250) \
+ geom_density(color='black', size=4, alpha=0) + scale_fill_brewer(type='seq')
```

```python
ggplot(data, aes(x='rating', fill='cond')) + ggsize(500, 250) \
+ geom_density(color='black', size=4, alpha=0.5) + scale_fill_brewer(type='seq')
```

```python
ggplot(data, aes(x='rating')) + ggsize(500, 250) \
+ geom_density(color='black', fill='blue', size=4, alpha=1)
```

```python
ggplot(data, aes(x='rating')) + ggsize(500, 250) \
+ geom_density(color='black', fill='blue', size=4, alpha=0.5)
```

```python
ggplot(data, aes(x='rating')) + ggsize(500, 250) + \
    geom_density(color='black', fill='blue', size=4, alpha=0)
```
