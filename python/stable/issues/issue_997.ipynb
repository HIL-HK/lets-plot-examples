The 'labels' parameter shouldn't affect custom ordering

https://github.com/JetBrains/lets-plot/issues/997

```python

data = {
    'g': ['b', 'a', 'a', 'a', 'c', 'c'],
}
labels = {'a': 'A', 'b': 'B', 'c': 'C'}

ggplot(data) + \
    geom_bar(aes(x=as_discrete('g', order_by='..count..'))) + \
    scale_x_discrete(labels=labels)
```
