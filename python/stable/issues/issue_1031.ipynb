the legend_justification parameter doesn't work if the legend_position='top' [closed]

https://github.com/JetBrains/lets-plot/issues/1031

```python
from lets_plot import *
LetsPlot.setup_html()
```

```python
data = {'x': [0, 1], 'y': [0, 1], 'g': ['A', 'B']}
ggplot(data, aes('x', 'y', color='g')) + \
    geom_point() + \
    theme(legend_position='top', legend_justification=[1, 1])
```

Should be


```python
ggplot(data, aes('x', 'y', color='g')) + \
    geom_point() + \
    theme(legend_position='bottom', legend_justification=[0, 1])
```

```python
ggplot(data, aes('x', 'y', color='g')) + \
    geom_point() + \
    theme(legend_position='left', legend_justification=[1, 1])
```

```python
ggplot(data, aes('x', 'y', color='g')) + \
    geom_point() + \
    theme(legend_position=[1,0], legend_justification=[1, 0])
```
