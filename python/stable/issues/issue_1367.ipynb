# [Display order of fill categories not being set correctly in stacked plots?](https://github.com/JetBrains/lets-plot/issues/1367)

```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
x = ["Q1"]*3 + ["Q2"]*3 + ["Q3"]*3
y = [9, 1, 1, 10, 7, 2, 11, 5, 1]
cat = ["A", "B", "C", "A", "C", "B", "A", "C", "B"]

dat = pd.DataFrame(
    dict(
        x=x,
        y=y,
        cat=cat
    )
)
```

```python
p1 = ggplot(dat) + geom_bar(aes('x', 'y', fill='cat'), stat='identity')
p1
```

`fill=as_discrete(levels..)` only changes order on the 'fill' color scale

```python
p2 = ggplot(dat) + geom_bar(aes('x', 'y', fill=as_discrete('cat', levels=["A", "C", "B"])), stat='identity')
p2
```

Pandas Categorical also only changes the order on the 'fill' color scale

```python
dat['cat'] = pd.Categorical(dat['cat'], categories=["A", "C", "B"], ordered=True)
p3 = ggplot(dat) + geom_bar(aes('x', 'y', fill='cat'), stat='identity')
p3
```

```python
# Expected

dat['cat'] = pd.Categorical(dat['cat'], categories=["A", "C", "B"], ordered=True)
dat = dat.sort_values('cat').reset_index(drop=True)

p4 = ggplot(dat) + geom_bar(aes('x', 'y', fill='cat'), stat='identity')

gggrid([
    p4 + ggtitle("Expected"),
    p2 + ggtitle("Actual")
])
```
