```python
import pandas as pd
from lets_plot import *

LetsPlot.setup_html()
```

`geom_path`: tooltips on a livemap feels more organically than on a regular plot [CLOSED]

https://github.com/JetBrains/lets-plot/issues/855

```python

minard_df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/minard/troops.csv')
minard_plot = geom_path(aes(x='long', y='lat', size='survivors', group='group', color='direction'), data=minard_df) \
    + scale_size([1.0, 20.0]) \
    + scale_color_manual(["#E1CBAE", "#232021"]) \
    + coord_map() \
    + flavor_solarized_light() \
    + guides(color = "none")


livemap = ggplot() \
    + geom_livemap() \
    + minard_plot

regular = ggplot() \
    + minard_plot

gggrid([livemap, regular]) + ggsize(1200, 400)
```
