# [facet_wrap does not skip empty groups](https://github.com/JetBrains/lets-plot/issues/1322)

```python
import pandas as pd
import numpy as np

from lets_plot import *
LetsPlot.setup_html()

# Input dataframe
df = pd.DataFrame(
    data = dict(
        c1 = ['a']*5 + ['b']*5,
        c2 = ['C']*2 + ['D']*3 + ['C']*5,
        i = list(np.arange(0,2)) + list(np.arange(0,3)) + list(np.arange(0,5)),
        val = np.arange(10,20)
    )
)

# Expected: 3 facets, not 4
( 
    ggplot(df)+
    aes(x='i', y='val')+
    geom_point()+
    facet_wrap(['c1', 'c2'])+
    ggsize(400,450)
)
```
