https://github.com/JetBrains/lets-plot/issues/553 [closed]

`boxplot`, `violin`, `crossbar`: position dodge width=0.95 should be used by default.

```python
np.random.seed(123)
data = DataFrame(dict(
    cond=np.repeat(['A','B'], 200),
    rating=np.concatenate((np.random.normal(0, 1, 200), np.random.normal(.8, 1, 200)))
))

p10 = (ggplot(data, aes(y='rating', fill="cond")) 
       + ggsize(300, 200) 
       + xlim(-.5, .5)
       + theme(legend_position='none'))

gggrid([p10 + geom_boxplot(), 
        p10 + geom_boxplot(position="dodge"), 
        p10 + geom_boxplot(position=position_dodge()), 
        p10 + geom_boxplot(width=.2), 
        p10 + geom_boxplot(width=.2, position="dodge"), 
        p10 + geom_boxplot(width=.2, position=position_dodge())], ncol=3)
```

```python
gggrid([p10 + geom_violin(), 
        p10 + geom_violin(position="dodge"), 
        p10 + geom_violin(position=position_dodge()), 
        p10 + geom_violin(width=.2), 
        p10 + geom_violin(width=.2, position="dodge"), 
        p10 + geom_violin(width=.2, position=position_dodge())], ncol=3)
```
