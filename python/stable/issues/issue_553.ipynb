https://github.com/JetBrains/lets-plot/issues/553

`boxplot`, `violin`, `crossbar`: position dodge width=0.95 should be used by default.

```python
np.random.seed(123)
data = DataFrame(dict(
    cond=np.repeat(['A','B'], 200),
    rating=np.concatenate((np.random.normal(0, 1, 200), np.random.normal(.8, 1, 200)))
))

p10 = (ggplot(data, aes(y='rating', fill="cond")) 
       + ggsize(300, 200) 
       + xlim(-.5, .5)
       + theme(legend_position='none'))
bunch = GGBunch()
bunch.add_plot(p10 + geom_boxplot(), 0, 0)
bunch.add_plot(p10 + geom_boxplot(position="dodge"), 300, 0)
bunch.add_plot(p10 + geom_boxplot(position=position_dodge()), 600, 0)

bunch.add_plot(p10 + geom_boxplot(width=.2), 0, 300)
bunch.add_plot(p10 + geom_boxplot(width=.2, position="dodge"), 300, 300)
bunch.add_plot(p10 + geom_boxplot(width=.2, position=position_dodge()), 600, 300)

bunch
```
