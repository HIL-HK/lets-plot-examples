`geom_histogram`: wrong plot area when `y='..density..'` [CLOSED]

https://github.com/JetBrains/lets-plot/issues/1157

```python
from lets_plot import *
LetsPlot.setup_html()

ggplot({'x': [-.5, 0, 0, .5]}, aes(x='x')) + geom_histogram(aes(y='..density..'), binwidth=.5)

# The plot area is expected to be 1, but it was 2.
```

```python
data = {
    'x': [0] * 4 + [2, 2] * 4 + [4] * 4,
    'g': ["A", "B"] * 8,
    't': ["t1", "t1", "t2", "t2"] * 4,
}
```

```python
ggplot(data, aes(x='x')) + geom_histogram(aes(y='..density..'), binwidth=2, center=0)
```

```python
ggplot(data, aes(x='x')) + geom_histogram(aes(y='..density..', fill='g'), binwidth=2, center=0)
```

```python
ggplot(data, aes(x='x')) + geom_histogram(aes(y='..density..'), binwidth=2, center=0) + facet_grid(x='g')
```

```python
ggplot(data, aes(x='x')) + geom_histogram(aes(y='..density..', fill='t'), binwidth=2, center=0) + facet_grid(x='g')
```
