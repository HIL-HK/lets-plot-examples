```python
LetsPlot.setup_html()
```

```python
data1 = {
    'x': [0, 1, 0, 1],
    'y': [0, 0, 1, 1],
    'fill': [0, 1, 1, 0],
}
```

```python
ggplot(data1) + geom_tile(aes('x', 'y', fill='fill'))
```

```python
ggplot(data1) + geom_livemap() + geom_tile(aes('x', 'y', fill='fill'))
```

```python
data2 = {
    'xmin': [0, 1, 0, 1],
    'ymin': [0, 0, 1, 1],
    'xmax': [1, 2, 1, 2],
    'ymax': [1, 1, 2, 2],
    'fill': [0, 1, 1, 0],
}
```

```python
ggplot(data2) + geom_rect(aes(xmin='xmin', xmax='xmax', ymin='ymin', ymax='ymax', fill='fill'))
```

```python
# Correct fill:
# |  blue | | black |
# | black | |  blue |
ggplot(data2) + geom_livemap() + geom_rect(aes(xmin='xmin', xmax='xmax', ymin='ymin', ymax='ymax', fill='fill'))
```
