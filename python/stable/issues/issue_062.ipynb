```python
mpg_df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
```

```python
ggplot(mpg_df, aes(y='hwy')) + geom_violin(trim=False)
```

```python
ggplot(mpg_df, aes(y='hwy')) + geom_violin()
```

```python
ggplot(mpg_df, aes(x=as_discrete('drv', order=1), y='hwy')) + \
    geom_violin(aes(fill='drv'), trim=False, alpha=.5)
```
