Integer numbers in facet strip titles are displayed as float

https://github.com/JetBrains/lets-plot/issues/1386 [closed]

Facet strip titles are displayed in different formats in Jupyter Notebook and when being exported to SVG - integer values got decimal part.

```python
from IPython.display import SVG
import pandas as pd
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg2.csv')

p = (
  ggplot(df, 
         aes('engine horsepower', 'miles per gallon', 
             color=as_discrete('origin of car',order=-1)))
  + geom_point()
  + theme_grey() + ggtitle('Efficiency vs Engine Horsepower')
)

q = p + facet_wrap(['origin of car', 'number of cylinders'], ncol=3)
q
```

```python
path_svg = ggsave(q, filename='facet_titles.svg')

SVG(filename=path_svg)
```
