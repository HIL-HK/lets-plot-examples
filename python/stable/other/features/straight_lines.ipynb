```python
ggplot() + \
    geom_vline(xintercept=2) + \
    geom_vline(xintercept=4, color='#006400', linetype=3, size=1.0) + \
    geom_vline(xintercept=6, color='#ff8000', linetype=2, size=2.0) + \
    geom_vline(xintercept=8, color='#ff00ff', linetype=4, size=3.0) + \
    geom_hline(yintercept=1) + \
    geom_hline(yintercept=3, color='magenta', linetype='dotdash', size=0.7) + \
    geom_hline(yintercept=5, color='dark_green', linetype='twodash', size=1.5) + \
    geom_hline(yintercept=7, color='gray', linetype='dashed', size=2.6) + \
    coord_fixed(ylim=[0, 8], xlim=[1, 9]) + \
    ggtitle('Horizontal and vertical lines of different style')
```

```python
ggplot() + \
    geom_abline(intercept=10, slope=-3.2, color='#5d8aa8', linetype='longdash', size=2) + \
    geom_abline(intercept=1, slope=-0.8, color='#5d8aa8', linetype='longdash', size=2) + \
    geom_abline(intercept=-1, slope=0.7, color='#5d8aa8', linetype='longdash', size=2) + \
    geom_abline(intercept=1, slope=3, color='#5d8aa8', linetype='longdash', size=2) + \
    geom_abline(intercept=3, slope=0, color='#5d8aa8', linetype='longdash', size=2) + \
    coord_fixed(ylim=[-3, 6], xlim=[-3, 6]) + \
    ggtitle('Sloping lines')
```
