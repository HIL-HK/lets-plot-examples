## Controlling Output Image Size with `w`, `h`, and `unit` in `ggsave()`

```kotlin
%use dataframe
%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
// Create a random density plot.

val rand = java.util.Random(123)
val n = 400
val data = mapOf (
    "rating" to List(n/2) { rand.nextGaussian() } + List(n/2) { rand.nextGaussian() * 1.5 + 1.5 },
    "cond" to List(n/2) { "A" } + List(n/2) { "B" }
)

var p = letsPlot(data) +
        geomDensity { x = "rating"; color = "cond" } + ggsize(500, 250)        
```

```kotlin
// A helper function to show the result of ggsave in a cell.

import java.io.File
import javax.imageio.ImageIO

fun showFile(path: String): Any {
    val file = File(path)
    val ext = file.extension.lowercase()

    return when (ext) {
        "svg", "htm", "html" -> HTML(file.readText())
        else -> {
            ImageIO.read(file)
        }
    }
}
```

### Export SVG to file

```kotlin
// By default the image inherits the plot size.
showFile(ggsave(p, "density.svg"))

// The result size is 500x250 pixels.
```

```kotlin
// You can override the original plot size by specifying `w` and `h` parameters. 
// For SVG, the default unit is "px" (pixels).
showFile(ggsave(p, "density_300_200.svg", w=300, h=200))

// The result size is 300x200 pixels.
```

```kotlin
// For "in", "mm", or "cm" units the actual size in pixels is defined by 96 dpi resolution.
// The `dpi` parameter is ignored for SVG.
showFile(ggsave(p, "density_4_2in.svg", w=4, h=2, unit="in"))

// The result size is 384x192 pixels; the physical size (4x2 in) is also saved in SVG attributes.
```

### Export HTML to file

```kotlin
// For HTML, `w` and `h` are always treated in pixels.
// The `unit` and `dpi` parameters are ignored.
showFile(ggsave(p, "density.html", w=250, h=120))
```

### Export PNG/JPG/TIFF to file

The file extension determines the raster format (PNG, JPG, or TIFF). 
Parameters behave the same across these formats; the examples here use PNG.

```kotlin
// With no size parameters, the default `scale=2` is applied to the original plot size (500x250).
showFile(ggsave(p, "density.png"))

// The result size is 1000x500 pixels.
```

```kotlin
// To get the original plot size, set `scale=1` explicitly.
showFile(ggsave(p, "density_scale1.png", scale=1))

// The result size is 500x250 pixels.
```

```kotlin
// For raster formats, the default unit is `"in"` and `dpi` is 300.
// When `w` and `h` are given, the default scale is 1.
showFile(ggsave(p, "density_3_2in.png", w=3, h=2))

// The result size is 900x600 pixels.
```

```kotlin
// Changing dpi affects the pixel dimensions for physical units.
// 4.5 in  3 in at 200 dpi produces the same pixel size as 3 in  2 in at 300 dpi.
// You can observe the difference by comparing the font size on the screen.

showFile(ggsave(p, "density_3_2in_dpi200.png", w=4.5, h=3, dpi=200))

// The result size is still 900x600 pixels.
```

```kotlin
// Using centimeters as the unit.
// 6 cm  3 cm = 2.36 in  1.18 in  multiplied by 200 dpi.
showFile(ggsave(p, "density_3_2cm.png", w=6, h=3, unit="cm", dpi=200))

// The result size is 473x238 pixels.
```

```kotlin
// To specify the exact pixel size, set `unit = "px"`.
showFile(ggsave(p, "density_300_200.png", w=300, h=200, unit="px"))

// The result size is 300x200 pixels.
```

```kotlin
// The `scale` parameter is applied after size and dpi calculations.
showFile(ggsave(p, "density_300_200_x2.png", w=300, h=200, unit="px", scale=2))

// The result size is 600x400 pixels.
```
