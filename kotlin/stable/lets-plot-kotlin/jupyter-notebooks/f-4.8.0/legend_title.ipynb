# Legend title

The legend title can be specified in `guideLegend()` and `guideColorbar()`.

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```kotlin
letsPlot(mpg.toMap()) + 
    geomPoint(size = 4) { x = "displ"; y = "hwy"; color = "cty"; shape = "cyl" } +
    guides(shape = guideLegend("Cylinders", ncol = 2),
           color = guideColorbar("City mileage"))
```
