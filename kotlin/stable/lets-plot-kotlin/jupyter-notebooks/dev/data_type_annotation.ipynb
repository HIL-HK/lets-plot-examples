# Default formatter based on data type

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
import java.time.Instant
import java.time.LocalDateTime
import java.time.Month
import java.time.ZoneOffset

val d = mapOf(
    "I" to listOf(1),
    "F" to listOf(1.2),
    "T" to listOf(LocalDateTime.of(2003, Month.FEBRUARY, 1, 0, 0, 0).toInstant(ZoneOffset.UTC))
)

val p = letsPlot(d) + 
    geomLabel(x=0, y=0) { label = "I" } +
    geomLabel(x=0, y=1) { label = "F" } +
    geomLabel(x=0, y=2) { label = "T" }
p
```

## LP-626
Int should be formatted without fractional part

```kotlin
val d = mapOf(
    "X" to listOf(0, 1, 2),
    "Y" to listOf(0, 0, 0),
    "I" to listOf(1, 2, 3)
)

val p = letsPlot(d) + geomLabel() { x = "X"; y = "Y"; label = "I" }
p
```

```kotlin
%use lib-ext
import java.nio.file.Files
import java.nio.file.Paths
val path = ggsave(p, "LP-626.png", scale=1)
Image(Files.readAllBytes(Paths.get(path)), "png")
```
