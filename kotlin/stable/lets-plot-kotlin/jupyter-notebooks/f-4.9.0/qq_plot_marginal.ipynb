# Parameter `marginal` of the Q-Q Plot

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val df = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
val dataMap = df.toMap()
df.head()
```

## Default Plot

```kotlin
gggrid(listOf(
    qqPlot(dataMap, sample = "hwy") + ggtitle("Q-Q Plot"),
    qqPlot(dataMap, x = "cty", y = "hwy") + ggtitle("Q-Q-2 Plot")
))
```

```kotlin
val marginals = listOf("none", "hist:tr", "dens:tr,hist:bl", "box : tr : .05, dens : bl")
gggrid(marginals.map { marginal ->
    qqPlot(dataMap, sample = "hwy", distribution = "exp", marginal=marginal) +
        ggtitle("marginal = '${marginal}'")
}, ncol = 2)
```
