# Line Type for Plot Elements


The `linetype` parameter in `elementLine()` and `elementRect()` specifies the line type for the element using either text ("blank", "solid", "dashed", "dotted", "dotdash", "longdash", "twodash") or number (0, 1, 2, 3, 4, 5, 6).

```kotlin
%use dataframe
%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```kotlin
val mpgData = mpg.toMap()
```

```kotlin
val p = letsPlot(mpgData) + geomPoint() { x = "cty"; y = "hwy"; color = "drv" }
p
```

```kotlin
p + theme(
    panelGrid = elementLine(linetype = "dashed", color = "grey"),
    panelGridMinorY = elementLine(linetype = "dotted", size = 0.75, color = "grey"),
    legendBackground = elementRect(linetype = 1, size = 0.75, color = "grey")
)
```
