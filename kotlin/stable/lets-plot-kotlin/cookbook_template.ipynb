# Plotting Points


`geomPoint()` draws markers on a scatter plot using different colors, shapes and sizes of markers.

1. [`color` and `size`](#color)
2. [`shape`](#shape)  
 2.1 [Switching to Hollow Shapes](#hollow-shape)

```kotlin
%use dataframe

%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

> *For demo, if possible, use standard data from [lets-plot-docs](https://github.com/JetBrains/lets-plot-docs/tree/master/data) repository.*
>
> *For example:*

```kotlin
val df = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
df.head(3)
```

> *The 1-st figure is usually built with a minimal set of required parameters.*
>
> *In many cases chapter header is not necessary.*

```kotlin
letsPlot(df.toMap()) + geomPoint {
    x = "displ"
    y = "hwy"
}
```

<a id="color"></a>
#### 1. `color` and `size`

```kotlin
letsPlot(df.toMap()) + geomPoint {
    x = "displ"
    y = "hwy"
    color = "drv"
    size = "displ"
}
```

<a id="shape"></a>
#### 2. `shape`

```kotlin
letsPlot(df.toMap()) + geomPoint(size = 8) {
    x = "displ"
    y = "hwy"
    shape = "drv"
}
```

<a id="hollow-shape"></a>
##### 2.1 Switching to Hollow Shapes 

```kotlin
letsPlot(df.toMap()) + geomPoint(size = 8) {
    x = "displ"
    y = "hwy"
    shape = "drv"
} + scaleShape(
        solid = false
    )
```
