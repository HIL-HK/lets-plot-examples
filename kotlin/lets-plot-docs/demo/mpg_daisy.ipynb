# MPG Daisy

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
val df = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
df.head()
```

```kotlin
letsPlot(df.toMap()) +
    geomBar(stat = Stat.identity, position = positionDodge()) { x = "model"; y = "cty"; fill = "cty" } +
    scaleFillBrewer(limits = 5 to 40, palette = "YlOrRd") +
    theme(
        axisLineY = elementLine(size = 2),
        axisLineX = elementLine(size = 2),
        axisTicksLengthY = 5,
        axisTicksLengthX = 10,
        axisTicksY = elementLine(size = 5),
        axisTicksX = elementLine(size = 3),
        axisTextX = elementText(angle = 10, vjust = .7, hjust = .4),
        panelInset = listOf(20, 140, 30, 135)
    ) +
    ggsize(900, 500) +
    coordPolar(transformBkgr = false)
```
