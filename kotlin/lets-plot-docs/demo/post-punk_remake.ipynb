# A post-punk chart remake

Original: https://blog.datawrapper.de/weekly-ridgeline-plot/

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
fun prepareData(df: DataFrame<*>): Map<String, List<*>> {
    val xs = mutableListOf<Int>()
    val ys = mutableListOf<Int>()
    val hs = mutableListOf<Double>()
    for (i in 0 until df.rowsCount()) {
        for (j in 0 until df.columnsCount()) {
            ys.add(i)
            xs.add(j)
            hs.add(df[j.toString()][i] as Double)
        }
    }
    val hAbsMin = abs(hs.min())
    return mapOf("x" to xs, "y" to ys, "h" to hs.map { it + hAbsMin })
}
```

```kotlin
val dataset = prepareData(DataFrame.readCSV(
    "https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/pulsar.csv",
    header = List(300) { it.toString() }
))
val rowsCount = dataset["h"]!!.size
```

```kotlin
letsPlot(dataset) +
    geomAreaRidges(stat = Stat.identity, scale = 0.25,
                   color = "white", fill = "black",
                   sampling = samplingPick(rowsCount))
        { x = "x"; y = "y"; height = "h" } +
    scaleYContinuous(trans = "reverse") +
    ggsize(600, 600) +
    themeVoid() +
    flavorHighContrastDark()
```
