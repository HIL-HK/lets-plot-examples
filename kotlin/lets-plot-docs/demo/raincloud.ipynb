# Raincloud Plot

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val df = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/refs/heads/master/data/penguins.csv")
val dataMap = df.toMap()
df.head()
```

```kotlin
letsPlot(dataMap) { x = "species"; y = "bill_depth_mm";
                    color = "species"; fill = "species" } +
    geomViolin(showHalf = 1, alpha = 0.75, trim = false,
               position = positionNudge(x = 0.07),
               tooltips = layerTooltips().format("@..violinwidth..", ".3~g")) +
    geomSina(showHalf = -1, fill = "white", shape = 21, seed = 42, showLegend = false,
             position = positionNudge(x = -0.07)) +
    geomBoxplot(width = 0.1, outlierAlpha = 0, showLegend = false) +
    scaleColorBrewer(palette = "Set2") +
    scaleFillBrewer(palette = "Pastel2") +
    coordFlip() +
    ylab("bill depth (mm)") +
    themeLight() +
    flavorSolarizedDark() +
    ggsize(800, 600)
```
