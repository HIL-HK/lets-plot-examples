# `statSummary()`

```kotlin
%use dataframe
%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
var mpgDf = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-kotlin/master/docs/examples/data/mpg.csv")
val mpgData = mpgDf.toMap()
mpgDf.head()
```

```kotlin
letsPlot(mpgData) { x = "drv"; y = "hwy" } + statSummary()
```

#### 1. The `geom` Parameter

```kotlin
letsPlot(mpgData) { x = "drv"; y = "hwy" } + statSummary(geom = Geom.crossbar())
```

#### 2. The `fn` Parameter

```kotlin
letsPlot(mpgData) { x = "drv"; y = "hwy" } + statSummary(geom = Geom.bar(), fn = "count")
```

#### 3. The `quantiles` Parameter

```kotlin
letsPlot(mpgData) { x = "drv"; y = "hwy" } +
    statSummary(fn = "mq", fnMin = "lq", fnMax = "uq", quantiles = listOf(0.45, 0.5, 0.55))
```
