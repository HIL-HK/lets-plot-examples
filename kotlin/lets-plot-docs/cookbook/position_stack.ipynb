```kotlin
%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val data = mapOf(
    "x" to listOf(1, 1, 2, 2, 2),
    "y" to listOf(1, 3, 2, 1, -3),
    "grp" to listOf("a", "b", "a", "b", "c")
)

val p = letsPlot(data) { x = "x"; y = "y"; group = "grp" } + 
    geomBar(stat = Stat.identity, color = "pen") { fill = "grp" }  + 
    scaleFillBrewer(palette = "Pastel1") +
    theme(axisLineX = elementBlank(), axisTicks = elementBlank()) + 
    geomHLine(yintercept = 0, color = "pen", size = 0.5)
```

```kotlin
val w = 600
val h = 400

gggrid(listOf(
    p + geomLabel(position = positionStack()) { label = "y" } + ggtitle("Default"),
    p + geomLabel(position = positionStack(0.0)) { label = "y" } + ggtitle("vjust = 0.0"),
    p + geomLabel(position = positionStack(0.3)) { label = "y" } + ggtitle("vjust = 0.3"),
    p + geomLabel(position = positionStack(0.5)) { label = "y" } + ggtitle("vjust = 0.5"),
    p + geomLabel(position = positionStack(0.7)) { label = "y" } + ggtitle("vjust = 0.7"),
    p + geomLabel(position = positionStack(1.0)) { label = "y" } + ggtitle("vjust = 1.0")
), ncol = 2) + ggsize(2 * w, 3 * h)
```

```kotlin
// position = "fill"

letsPlot(data) { x = "x"; y = "y"; group = "grp" } + 
    geomBar(stat = Stat.identity, position = positionFill(), color = "pen") { fill = "grp" }  + 
    scaleFillBrewer(palette = "Pastel1") +
    geomText(position = positionFill(0.5)) { label = "y" } +
    theme(axisLineX = elementBlank(), axisTicks = elementBlank()) + 
    geomHLine(yintercept = 0, color = "pen", size = 0.5)
```
