# Flipping Coordinates

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
var mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-kotlin/master/docs/examples/data/mpg.csv")
mpg.head(3)
```

## `coordFlip()`

```kotlin
val p = letsPlot(mpg.toMap()) + theme(axisTitleX="blank")
```

```kotlin
val boxPlot = p + 
    geomBoxplot(color="#579673", fill="#9AC0B3", size=1.5) {
        x = asDiscrete("class", order=1, orderBy="..middle..")
        y = "hwy" 
    } + 
    ggsize(750, 300)
    
boxPlot + ggtitle("Default")    
```

```kotlin
boxPlot + ggtitle("Flipped") + coordFlip()
```

```kotlin
val barPlot = p + 
    geomBar(size=1.5, alpha=0.7, width=0.7) {
        x = asDiscrete("manufacturer", orderBy="..count..")
        fill = "class"
        color = "class"
    } +
    ggsize(800, 300)
    
barPlot + ggtitle("Default")
```

```kotlin
barPlot + ggtitle("Flipped") + coordFlip()
```

## Parameter `flip`

```kotlin
val pointPlot = letsPlot(mpg.toMap()) +
    geomPoint(alpha = .5, shape = 21,
              position = positionJitter(width = .2, height = .2, seed = 42))
        { x = "displ"; y = "hwy"; fill = "class" }

pointPlot + coordFixed(ratio = .1) + ggtitle("coordFixed()")
```

```kotlin
pointPlot + coordFixed(ratio = 10, flip = true) + ggtitle("coordFixed(flip = true)")
```
