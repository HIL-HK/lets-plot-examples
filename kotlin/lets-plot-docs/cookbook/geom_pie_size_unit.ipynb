# Parameter `sizeUnit` in `geomPie()`




Use the `sizeUnit` parameter to relate the size of the "pie"\
to the length of the unit step along one of the axis.


```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```kotlin
letsPlot(mpg.toMap()) + geomPie {
    fill = "class"
    weight = "displ"
}
```

#### 1. Make Pie' Diameter Equal to the Unit Step along the X-axis

Parameters: `size=1, sizeUnit="x"`.

```kotlin
letsPlot(mpg.toMap()) + geomPie(size = 1, sizeUnit = "x") {
    fill = "class"
    weight = "displ"
} + coordFixed()
```

#### 2.  Decrease the Pie' Diameter Slightly

Parameters: `size=0.7, sizeUnit="x"`.

```kotlin
letsPlot(mpg.toMap()) + geomPie(size = 0.7, sizeUnit = "x") {
    fill = "class"
    weight = "displ"
} + coordFixed() + themeVoid()
```
