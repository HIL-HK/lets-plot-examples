# Internet use and activities in Europe
Frequency of internet access: daily

Unit of measure: Percentage of individuals

Year: 2016

```python
df = pd.read_csv('https://raw.githubusercontent.com/HIL-HK/lets-plot-examples/master/data/internet_use_and_activities.csv')
df.head()
```

```python
boundaries_gpd = regions_country(request=df.country).boundaries()
```

```python
p = ggplot() + geom_livemap(zoom=4, location=[10, 53]) + \
    geom_polygon(mapping=aes(fill='percent'), data=df, map=boundaries_gpd, map_join=['country', 'request'], 
                 alpha=0.6, color='black', size=0.1)
p + scale_fill_gradient(high='red')
```
