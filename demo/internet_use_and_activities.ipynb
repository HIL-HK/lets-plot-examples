# Internet Use and Activities in Europe

Frequency of internet access: daily

Unit of measure: Percentage of individuals

Year: 2016

```python
import pandas as pd

from lets_plot import *
from lets_plot.geo_data import *
from lets_plot.settings_utils import maptiles_lets_plot
from lets_plot.settings_utils import geocoding_service
LetsPlot.set(geocoding_service(url='https://geo.datalore.jetbrains.com'))
LetsPlot.setup_html()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com'
                 '/HIL-HK/lets-plot-examples/master/data/internet_use_and_activities.csv')
df.head()
```

```python
boundaries_gpd = regions_country(request=df.country).boundaries()
```

```python
ggplot() + \
    geom_livemap(zoom=4, location=[10, 53]) + \
    geom_polygon(aes(fill='percent'), data=df, map=boundaries_gpd, map_join=['country', 'request'], \
                 alpha=0.6, color='black', size=0.1) + \
    scale_fill_gradient(high='red') + \
    ggsize(800, 600) + \
    ggtitle('Internet Use and Activities in Europe')
```
